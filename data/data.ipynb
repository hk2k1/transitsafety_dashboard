```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('juneroute.csv')           #imports the june route data
df['Time'] = pd.to_datetime(df['Time'])     #use date time to extract 
```

```python
df['day'] = df['Time'].apply(lambda x:x.day) #brings out another column with only the day
df
```

```python
df_events = df.groupby(by = 'day', as_index = False).count()[['day', 'Time']].rename(columns = {'Time': 'event count'})     #groups the data according to day and also counts the diff events using time
df_events
```

```python
df_events['highest speed'] = df.groupby('day', as_index = False).max()['Speed']     #adds column for lowest and max speed for each day
df_events['lowest speed'] = df.groupby('day', as_index = False).min()['Speed']
df_events
```
